5. **Cải Thiện Thương Hiệu:**

# 4. Phân tích lí do trong số 5000 khách hàng này, chỉ có 480 (= 9,6%) chấp nhận khoản vay cá nhân

```python
# Vẽ biểu đồ countplot
plt.figure(figsize=(12, 6))
sns.countplot(x='Age', data=Bank_Personal_Loan_Modelling, hue='Personal Loan')

# Tính toán và hiển thị con số
grouped_data_age = Bank_Personal_Loan_Modelling.groupby(['Age', 'Personal Loan']).size().reset_index(name='Count')

for index, row in grouped_data_age.iterrows():
    age_group = row['Age']
    loan_acceptance = 'Accepted' if row['Personal Loan'] == 1 else 'Not Accepted'
    loan_count = row['Count']
    
    print(f"Age Group {age_group}: {loan_acceptance} Personal Loan - Count: {loan_count}")

plt.show()
```

- Có thể thấy rằng độ tuổi 23, 24, 25 và 66, 67 thì khách hàng không chấp nhận khoản vay cá nhân, ví dụ như ở độ tuổi 25 trong 53 người thì không có lấy 1 người chấp nhận, điều này cũng cần lưu ý

```python
# Vẽ biểu đồ countplot
plt.figure(figsize=(12, 6))
sns.countplot(x='Experience', data=Bank_Personal_Loan_Modelling, hue='Personal Loan')

# Tính toán và hiển thị con số
grouped_data_exp = Bank_Personal_Loan_Modelling.groupby(['Experience', 'Personal Loan']).size().reset_index(name='Count')

for index, row in grouped_data_exp.iterrows():
    exp_group = row['Experience']
    loan_acceptance = 'Accepted' if row['Personal Loan'] == 1 else 'Not Accepted'
    loan_count = row['Count']
    
    print(f"Experience Group {exp_group}: {loan_acceptance} Personal Loan - Count: {loan_count}")

plt.show()
```

- Có thể thấy rằng khi số năm kinh nghiệm là 42 và 43 năm thì có xu hướng không chấp nhận khoản vay cá nhân

```python
data_X = Bank_Personal_Loan_Modelling.loc[:, Bank_Personal_Loan_Modelling.columns  != 'Personal Loan']
data_Y = Bank_Personal_Loan_Modelling[['Personal Loan']]
```

```python
pt = PowerTransformer(method='yeo-johnson',standardize=False)
pt.fit(data_X['Income'].values.reshape(-1,1))
temp = pt.transform(data_X['Income'].values.reshape(-1,1))
data_X['Income'] = pd.Series(temp.flatten())

# Distplot to show transformed Income variable
sns.distplot(data_X['Income'])
plt.show()
```

```python
# Applying the Yeo Johnson method of Transformation on the CCAvg variable.
pt = PowerTransformer(method='yeo-johnson',standardize=False)
pt.fit(data_X['CCAvg'].values.reshape(-1,1))
temp = pt.transform(data_X['CCAvg'].values.reshape(-1,1))
data_X['CCAvg'] = pd.Series(temp.flatten())

# Distplot to show transformed CCAvg variable
sns.distplot(data_X['CCAvg'])
plt.show()
```

```python
# Binning on Mortgage variable.
data_X['Mortgage_Int'] = pd.cut(data_X['Mortgage'],
                               bins=[0,100,200,300,400,500,600,700],
                               labels= [0,1,2,3,4,5,6],
                               include_lowest =True)
data_X.drop('Mortgage', axis = 1, inplace= True)
```

```python
# Splitting the data into train and test. We use stratify parameter of train_test_split function to get the same class distribution across train and test sets.
X_train,X_test,Y_train,Y_test = train_test_split(data_X,data_Y,test_size = 0.3, random_state = 0,stratify = data_Y)

X_train.reset_index(drop= True, inplace= True);
X_test.reset_index(drop= True, inplace= True);
Y_train.reset_index(drop= True, inplace= True);
Y_test.reset_index(drop= True, inplace= True);

# To display top 5 rows
X_train.head()
```

```python
# We will apply the StandardScaler to the dataset to standardize the input variables

for ind, column in enumerate(X_train.columns):
    scaler = StandardScaler()
    
    #fit to train data
    scaler.fit(X_train[[column]])
    
    #transform train data
    np_array = scaler.transform(X_train[[column]])
    X_train.loc[: , column] = pd.Series(np_array.flatten())
    
    #transform test data
    np_array = scaler.transform(X_test[[column]])
    X_test.loc[: , column] = pd.Series(np_array.flatten())
    
```

## Logistic Regression

```python
# Xử lý giá trị thiếu bằng cách điền giá trị trung bình
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)

# Tạo và đào tạo mô hình
model = LogisticRegression(random_state=0)
model.fit(X_train, Y_train)
```

```python
X_test_pred1 = model.predict(X_test)

# Accuracy of train data
train_accuracy = model.score(X_train, Y_train)
print(f"Accuracy on Train Data: {train_accuracy:.2%}")

# Accuracy of test data
test_accuracy = model.score(X_test, Y_test)
print(f"Accuracy on Test Data: {test_accuracy:.2%}")

# Confusion Matrix
def plot_confusion_matrix(actual, predicted):
    cm = confusion_matrix(actual, predicted)
    fig, ax = plt.subplots(figsize=(8,6))
    ax.set_ylim([0,5])
    sns.heatmap(cm, annot=True, fmt= '.2f', xticklabels=[0, 1], yticklabels=[0, 1])
    plt.ylabel('Observed')
    plt.xlabel('Predicted')
    plt.title('Confusion Matrix')
    plt.show()

print('Confusion Matrix on Test Data:')
plot_confusion_matrix(Y_test, X_test_pred1)

# Additional Metrics
precision = precision_score(Y_test, X_test_pred1)
recall = recall_score(Y_test, X_test_pred1)
f1 = f1_score(Y_test, X_test_pred1)
roc_auc = roc_auc_score(Y_test, X_test_pred1)

print(f"\nPrecision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print(f"ROC AUC: {roc_auc:.2f}")
```

# 6. Phân loại khách hàng

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Chuẩn bị dữ liệu
selected_columns = ['Age', 'Income', 'CCAvg', 'Education']
data_for_clustering = Bank_Personal_Loan_Modelling[selected_columns]

# Sử dụng phương pháp "elbow" để chọn số lượng cụm tối ưu
inertia = []

# Thử nghiệm số lượng cụm từ 1 đến 10 (hoặc một giới hạn nào đó)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_for_clustering)
    inertia.append(kmeans.inertia_)

# Hiển thị biểu đồ "elbow"
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Sử dụng thuật toán K-Means với số lượng cụm (k) là 2
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(data_for_clustering)

# Thêm thông tin về cụm vào DataFrame
Bank_Personal_Loan_Modelling['Cluster'] = clusters

# Hiển thị thông tin cụm
print(Bank_Personal_Loan_Modelling.groupby('Cluster').mean())
```
